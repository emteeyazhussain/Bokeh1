```python
#Q1.
# create a Bokeh plot using Python code with the following steps:

from bokeh.plotting import figure, show

# Prepare the data
x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

# Create a figure
p = figure(plot_width=400, plot_height=400, title='My Bokeh Plot')

# Add glyphs to the figure
p.circle(x, y)

# Display the plot
show(p)
```

```python
#Q2.

# In Bokeh, a glyph is a graphical representation of data. Glyphs can be thought of as the visual building blocks of Bokeh plots. There are several 
# types of glyphs available in Bokeh, including circles, squares, lines, bars, and more. Each glyph type has its own set of attributes that can be 
# customized to control the appearance of the glyph.

from bokeh.plotting import figure, show

# create some data
x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

# create a new plot with default tools, using figure
p = figure(plot_width=400, plot_height=400)

# add a circle glyph with size, color, and alpha attributes
p.circle(x, y, size=20, color="navy", alpha=0.5)

# show the plot
show(p)
```

```python
#Q3.
from bokeh.plotting import figure, show
from bokeh.models import Legend, LegendItem

# create a new plot with a title and axis labels
p = figure(title="My Plot", x_axis_label="X-axis label", y_axis_label="Y-axis label")

# customize the plot background and borders
p.background_fill_color = "beige"
p.border_fill_color = "gray"
p.min_border_left = 50
p.min_border_right = 50
p.min_border_top = 50
p.min_border_bottom = 50

# customize the title
p.title.text_font_size = "18pt"
p.title.text_color = "navy"

# customize the axis labels
p.xaxis.axis_label_text_font_size = "14pt"
p.xaxis.axis_label_text_color = "red"
p.yaxis.axis_label_text_font_size = "14pt"
p.yaxis.axis_label_text_color = "green"

# customize the tick labels
p.xaxis.major_label_text_font_size = "12pt"
p.xaxis.major_label_text_color = "blue"
p.yaxis.major_label_text_font_size = "12pt"
p.yaxis.major_label_text_color = "purple"

# add some data to the plot
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
p.line(x, y1, line_width=2, color="red", legend_label="Line 1")
p.line(x, y2, line_width=2, color="green", legend_label="Line 2")

# customize the legend
legend_items = [
    LegendItem(label="Line 1", renderers=[p.renderers[0]]),
    LegendItem(label="Line 2", renderers=[p.renderers[1]])
]
legend = Legend(items=legend_items, location="top_left")
p.add_layout(legend, "right")

# show the plot
show(p)
```

```python
#Q5.
# we need to follow the following steps:

# Step 1: Create a Bokeh Plot
# We can create a Bokeh plot using the Bokeh library in Python. Here's an example code to create a simple Bokeh plot:

from bokeh.plotting import figure
from bokeh.embed import components

# Create a Bokeh figure object
plot = figure(title='Example Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add some data to the plot
plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])

# Get the Bokeh components of the plot
script, div = components(plot)

# Step 2: Save the Bokeh Plot as an HTML File
from bokeh.plotting import output_file, save

# Set the output file name
output_file("example_plot.html")

# Save the plot to the output file
save(plot)


# Step 3: Serve the HTML File using Flask or Django

from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    # Render the HTML template with the Bokeh components
    return render_template("index.html", script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)


# We can use similar code in Django to embed a Bokeh plot into a web page or dashboard. In Django, we would create a view function that renders the HTML
#  template, and define a URL pattern to map the view function to a URL. The HTML template would be similar to the one used in Flask, but with
#   Django-specific syntax for including the Bokeh components








```
